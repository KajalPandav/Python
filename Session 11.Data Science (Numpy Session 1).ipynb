```python
## Reshape (Change the dimension of array) ##

import numpy as ac
ar1=ac.array([65,77,45,67,34,56,23,45])
ar1.reshape(4,2)           ##(4=Rows,2=Columns)##
```

```python
import numpy as ac
ar1=ac.array([65,77,45,67,34,56,23,45])
ar1.reshape(2,4)    ##(2=Rows,4=Columns)##
```
